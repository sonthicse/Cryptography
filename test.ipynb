```python
# AFFINE
def affine_encrypt(plaintext, k1, k2):
    plaintext = plaintext.upper()
    ciphertext = ""
    offset = ord('A')
    for char in plaintext:
        encrypted_char = chr(((k1 * (ord(char) - offset) + k2) % 26) + offset)
        ciphertext += encrypted_char
    return ciphertext
```

```python
plaintext = "HELLO"
k1 = 5
k2 = 8
ciphertext = affine_encrypt(plaintext, k1, k2)
print("Ciphertext:", ciphertext)
```

```python
def egcd(r1, r2):
    s1, s2 = 1, 0
    t1, t2 = 0, 1
    while r2 > 0:
        q = r1 // r2
        r = r1 - q * r2
        s = s1 - q * s2
        t = t1 - q * t2
        r1, r2 = r2, r
        s1, s2 = s2, s
        t1, t2 = t2, t
    return r1, s1, t1

print(egcd(48, 16))
    
```
