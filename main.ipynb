## Ceasar Cipher

```python
P = "hocvienkythuatquansu"
K = 10
C = ''

# Iterate over each character in string P
def E(P, K):
	C = ''
	for i in range(0, len(P), 1):
		convert_alphabet = ord(P[i]) - ord('a')
		new_character = (convert_alphabet + K) % 26
		convert_ascii = new_character + ord('a')
		C = C + chr(convert_ascii)
	return C

def D(C, K):
	P = ''
	for i in range(0, len(C), 1):
		convert_alphabet = ord(C[i]) - ord('a')
		new_character = ((convert_alphabet - K) % 26 + 26) % 26
		convert_ascii = new_character + ord('a')
		P = P + chr(convert_ascii)
	return P

C = E(P, K)
print(C)

P = D(C, K)
print(P)
```
