## Sinh khóa

Chọn 2 số nguyên tố p và q lớn ngẫu nhiên <br>
Điều kiện:
- p và q không được quá gần nhau ($p \approx q$)
- $p/q$ không được gần các số hữu tỉ đẹp

```python
bit = 1024
p = number.getPrime(bit)
q = number.getPrime(bit)
while True:
    p = number.getPrime(bit)
    q = number.getPrime(bit)
    if abs(p - q).bit_length() > bit // 2:
        break

print("p =", p)
print("q =", q)
```

Tính $n=p \cdot q$

```python
n = p * q
print("n =", n)
```

Tính $\phi(n) = (p - 1) \cdot (q - 1)$

```python
phi = (p - 1) * (q - 1)
print("φ(n) =", phi)
```

Chọn e sao cho $1 < e < φ(n)$ và $\gcd(e, φ(n)) = 1$

```python
e = 65537
```

Tìm nghịch đảo modulo của e
$$d \cdot e = 1 \mod{\phi(N)}$$

```python
d = number.inverse(e, phi)
print("d =", d)
```

```python
print("Public key: (e =", e, ", n =", n, ")")
print("Private key: (d =", d, ", n =", n, ")")

with open("public_key.txt", "w") as f:
    f.write(f"{e}\n{n}\n")
with open("private_key.txt", "w") as f:
    f.write(f"{d}\n{n}\n")
```

```python
with open("plaintext.txt", "r") as f:
	plaintext = f.read()

print("Độ dài bản rõ:", len(plaintext))
print("Bản rõ:", plaintext)
```

```python
M = [ord(c) for c in plaintext]
print(M)
```

## Mã hóa

$$C = P^e \mod{N}$$

```python
ciphertext = []

for i in range(len(M)):
	# input("Nhấn Enter để tiếp tục...")
	C = pow(M[i], e, n)
	ciphertext.append(C)
	print(f"M[{i}] = {M[i]} -> C[{i}] = {C}")
```

```python
print(type(ciphertext))
print("Ciphertext:", ciphertext)

with open("ciphertext.txt", "w") as f:
	for C in ciphertext:
		f.write(f"{C}\n")
```

## Giải mã

$$P = C^d \mod{N}$$

```python
decrypted = []

for i in range(len(ciphertext)):
    C = ciphertext[i]
    M_decrypted = pow(C, d, n)
    decrypted.append(M_decrypted)
    print(f"C[{i}] = {C} -> M[{i}] = {M_decrypted}")
```

```python
print("Decrypted:", decrypted)

with open("decrypted.txt", "w") as f:
	for M_decrypted in decrypted:
		f.write(f"{M_decrypted}\n")
```
