Import thư viện

```python
from Crypto.Cipher import AES
import os
```

Đọc khóa từ file `key.bin`

```python
with open("key.bin", "rb") as f:
    key = f.read()
    
print("Độ dài khóa:", len(key))
print("Khóa:", key.hex())
print(key.decode("utf-8"))
```

Đọc bản rõ từ file `plaintext.txt`

```python
with open('plaintext.txt', 'rb') as f:
    plaintext = f.read()

print("Độ dài bản rõ:", len(plaintext))
print("Bản rõ:", plaintext.hex())
print(plaintext.decode("utf-8"))
```

Nếu độ dài bản rõ không chi hết cho 128 bit

```python
if (len(plaintext) % 16 != 0):
	padding = 16 - (len(plaintext) % 16)
	plaintext += bytes([padding] * padding)
	print("Độ dài bản rõ sau khi padding:", len(plaintext))
	print("Bản rõ sau khi padding:", plaintext.hex())
```

Tạo đối tượng mã hóa AES theo mode

```python
cipher_en = AES.new(key, AES.MODE_ECB)
ciphertext = b''
```

Mã hóa từng khối theo AES

```python

for i in range(0, len(plaintext), 16):
    block = plaintext[i:i+16]
    print("Khối bản rõ [{}]:".format(i//16 + 1), block.hex())
    input("Nhấn Enter để tiếp tục...")
    cipher_block = cipher_en.encrypt(block)
    print("Khối bản mã [{}]:".format(i//16 + 1), cipher_block.hex())
    ciphertext += cipher_block
```

```python
print("Độ dài bản mã:", len(ciphertext))
print("Bản mã:", ciphertext.hex())
print(ciphertext.decode("utf-8", errors='ignore'))
with open('ciphertext.bin', 'wb') as f:
	f.write(ciphertext)
```

Tạo đối tượng giải mã AES theo mode

```python
cipher_de = AES.new(key, AES.MODE_ECB)
decrypted = b''
```

```python
for i in range(0, len(ciphertext), 16):
    block = ciphertext[i:i+16]
    print("Khối bản mã [{}]:".format(i//16 + 1), block.hex())
    input("Nhấn Enter để tiếp tục...")
    plain_block = cipher_de.decrypt(block)
    print("Khối bản rõ [{}]:".format(i//16 + 1), plain_block.hex())
    decrypted += plain_block
```

Xóa padding sau khi giải mã

```python
decrypted = decrypted[:-decrypted[-1]]
```

```python
print("Độ dài bản mã:", len(decrypted))
print("Bản mã:", decrypted.hex())
print(decrypted.decode("utf-8", errors='ignore'))
with open('decrypted.txt', 'wb') as f:
	f.write(decrypted)
```
