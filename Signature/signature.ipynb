```python
from Crypto.PublicKey import RSA
from Crypto.Hash import SHA256
from Crypto.Util import number
```

### Tạo khóa RSA

```python
key = RSA.generate(2048)
print(key.p)
print(key.q)

with open("priv.pem", "wb") as f:
    f.write(key.export_key())

with open("pub.pem", "wb") as f:
    f.write(key.publickey().export_key())
```

```python
n = key.n
d = key.d
```

### Đọc file ảnh

```python
with open("avatar.jpeg", "rb") as f:
    data = f.read()
```

```python
h_bytes = SHA256.new(data).digest()
print(h_bytes)
h_int = int.from_bytes(h_bytes, "big")
print(h_int)
```

```python
sign_int = pow(h_int, d, n)
sign_bytes = number.long_to_bytes(sign_int)
print(sign_int)
```

### Ghép chữ ký vào file

```python
signed = data + sign_bytes

with open("sign.txt", "wb") as f:
    f.write(sign_bytes)

with open("signed.jpeg", "wb") as f:
    f.write(signed)
```

```python
with open("pub.pem", "rb") as f:
    pub_key = RSA.import_key(f.read())

n = pub_key.n
e = pub_key.e
```

```python
with open("avatar.jpeg", "rb") as f:
    M = f.read()
M_int = int.from_bytes(M, byteorder="big")

with open("sign.txt", "rb") as f:
    S = f.read()
S_int = int.from_bytes(S, byteorder="big")


H = SHA256.new(M).digest()
H_int = int.from_bytes(H, byteorder="big")
H_S = pow(S_int, e, n)

if H_S == H_int:
    print("OK")
else:
    print("Sai")
```
